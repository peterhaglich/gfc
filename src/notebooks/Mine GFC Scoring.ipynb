```python
for dd in the_days:
    print("Processing {}".format(dd))
    _created_after = "{0}T00:00:00".format(dd)
    _created_before = "{0}T23:59:59".format(dd)
    params = {"created_after": parse(_created_after),
              "created_before": parse(_created_before)}
    scores = gf.get_scores(**params)
    if scores is None:
        pass
    else:
        _path = os.path.join(SCORING_PATH,
                             "Scoring_Snapshot_{0}.json".format(dd))
        with open(_path, "w") as f:
            json.dump(scores, f, ensure_ascii=False, sort_keys=True, indent=2)
```
